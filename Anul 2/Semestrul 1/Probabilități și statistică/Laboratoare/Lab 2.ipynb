```python
from random import randint

def zile_nastere_sim(n, nr_sim = 10000):
    zile_nastere = []
    cnt_coincid = 0
    for i in range(nr_sim):
        zile_nastere.clear()
        for j in range(n):
            z = randint(1,365)
            zile_nastere.append(z)
        cnt_coincid += (len(set(zile_nastere)) != n)
    return cnt_coincid / nr_sim

zile_nastere_sim(30)
```

```python
from matplotlib.pyplot import plot, grid, title, show

title('Plot test')
xs = [5,10,15,20,25,29,35,40,45,50,57]
ys = [zile_nastere_sim(x) for x in xs]
plot(xs,ys,'r*')
grid()
show()
```

```python
from random import random
from matplotlib.pyplot import axis
from math import dist


axis('square')
axis((0,1,0,1))
centru = [0.5, 0.5]
for i in range(1000000):
    F = [random(), random()]
    if (dist(F,centru) > 0.5):
        plot(F[0],F[1],'r*')
    else:
        plot(F[0],F[1],'b*')
show()
```

```python
axis('square')
axis((0,1,0,1))
centru = [0.5, 0.5]
A = [0,0]
B = [1,1]
C = [1,0]
D = [0,1]
for i in range(10000):
    F = [random(), random()]
    if (dist(F,centru) < dist(F,A) and dist(F,centru) < dist(F,B) and dist(F,centru) < dist(F,C) and dist(F,centru) < dist(F,D)):
        plot(F[0],F[1],'b*')
    else:
        plot(F[0],F[1],'r*')
show()
```

```python
axis('square')
axis((0,1,0,1))
centru = [0.5, 0.5]
A = [0,0]
B = [1,0]
C = [1,1]
D = [0,1]
for i in range(100000):
    F = [random(), random()]
    ascutite = 0
    if(dist(F,A) * dist(F,A) + dist(F,B) * dist(F,B) > 1):
        ascutite += 1
    if(dist(F,B) * dist(F,B) + dist(F,C) * dist(F,C) > 1):
        ascutite += 1
    if(dist(F,C) * dist(F,C) + dist(F,D) * dist(F,D) > 1):
        ascutite += 1
    if(dist(F,A) * dist(F,A) + dist(F,D) * dist(F,D) > 1):
        ascutite += 1
    if(ascutite == 2):
        plot(F[0],F[1], 'b*')
    else:
        plot(F[0],F[1],'r*')
show()
```

```python
#Metoda teoretica

def zile_nastere_sim_teoretic(c):
    p = 1
    n = 0
    while(n < c):
        p *= (365 - n) / 365
        n += 1
    return 1 - p

zile_nastere_sim_teoretic(30)
```
