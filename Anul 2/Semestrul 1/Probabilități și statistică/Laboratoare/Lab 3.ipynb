```python
# 2. Se simuleaza aruncarea unui zar de 500 de ori. Afisati histograma frecventelor relative 
#    ale numerelor obtinute si barele corespunzatoare probabilitatilor teoretice, inlcouind 
#    semnele de intrebare din codul urmator.

from random import randrange
from matplotlib.pyplot import bar, hist, grid, show, legend

data = [randrange(1,7) for _ in range (500)]
bin_edges = [k+0.5 for k in range(7)]
hist(data, bin_edges, density = True, rwidth = 0.9, color = 'green', edgecolor = 'black', alpha = 0.5, label = 'frecvente relateive')

distribution = dict([(i,1/6) for i in range(1,7)])
bar(distribution.keys(), distribution.values(), width = 0.85, color = 'red', edgecolor = 'black', alpha = 0.6, label = 'probabilitati')

legend(loc = 'lower left')
grid()
show()
```

```python
# 3. Intr-o urna sunt 6 bile cu cifra 1 inscriptionata si 4 bile cu cifra 0. Se extrag aleator 
#    cu returnare 5 bile din urna. Fie X variabila aleatoare care indica suma numerelor de pe bilele extrase

from scipy.stats import binom


data = binom.rvs(n = 5, p = 6/10, size = 1000)
bin_edges = [k-0.5 for k in range(0,7)]
hist(data, bin_edges, density = True, rwidth = 0.9, color = 'green', edgecolor = 'black', alpha = 0.5, label = 'frecvente relateive')

distribution = dict([(i, binom.pmf(i, n = 5, p = 6/10)) for i in range(0,6)])
bar(distribution.keys(), distribution.values(), width = 0.85, color = 'red', edgecolor = 'black', alpha = 0.6, label = 'probabilitati')

legend(loc = 'lower left')
grid()
show()
```

```python
# 1. MontyHall

from random import sample

cnt = 0
for i in range(100000):
    asezare = list('' * 3)
    c1, c2, m = sample([0,1,2], k = 3)
    asezare[m] = ''
    # print('Asezare:          ', asezare)
    aleg1 = randrange(3)
    usi = list('' * 3)
    usi[aleg1] = ''
    # print('Prima alegere:    ', usi)
    deschid = ''
    aleg2 = ''
    if aleg1 == c1:
        deschid = c2
        aleg2 = m

    elif aleg1 == c2:
        deschid = c1
        aleg2 = m

    else:
        deschid = sample([c1, c2], k = 1)[0]
        if deschid == c1:
            aleg2 = c2
        else:
            aleg2 = c1

    usi[deschid] = ''
    # print('Capra dezvaluita: ', usi)

    usi[aleg1] = ''
    usi[aleg2] = ''
    usi[deschid] = ''
    # print('A doua alegere:   ', usi)

    if(aleg2 == m):
        cnt += 1
        # print("SUCCES")
    # else:
        # print("ESEC")
print((cnt * 100)//100000,'%')
```
