```python
from scipy.stats import norm
from numpy import mean, std, linspace
from matplotlib.pyplot import show, hist, grid, xticks, plot

m = 165
s = 10
gen = 10000

# cnt = 0
# for _ in range(gen):
#     data = norm.rvs(loc = m, scale = s, size = gen)
#     for _ in data:
#         if _ >= 200:
#             cnt += 1
# print((cnt / gen) / 100)

data = norm.rvs(loc = m, scale = s, size = gen)
hist(data, bins = 16, range = (130, 210), density = True)
xticks(range(130, 210, 5), fontsize = 8)
x = linspace(130,210,gen)
plot(x, norm.pdf(x, loc = m, scale = s))
print(mean(data))
print(std(data))
print(mean([160 < x < 170 for x in data]))
deviatie160 = norm.cdf(160, loc = m, scale = s)
deviatie170 = norm.cdf(170, loc = m, scale = s)
print(deviatie170 - deviatie160)
grid()
show()
```

```python
from scipy.stats import expon, uniform
from numpy import mean, std

def genereaza_print():
    y = uniform.rvs()
    t = 0
    if(y > 0.4):
        t = uniform.rvs(loc = 4, scale = 2)
    else:
        t = expon.rvs(loc = 0, scale = 5)
    t = float(t)
    return t

data = [genereaza_print() for _ in range(1000)]
print(mean(data))
print(std(data))
print(mean([x < 5 for x in data]))
print(0.4 * expon.cdf(5, loc = 0, scale = 5) + 0.6 * uniform.cdf(5, loc = 4, scale =2))
```
