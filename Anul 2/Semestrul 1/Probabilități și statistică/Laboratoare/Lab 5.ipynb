```python
from scipy.stats import uniform
from numpy import mean

Y = uniform.rvs(size = 100000)
mean([(0.25 < y < 0.5) or (0.75 < y < 1) for y in Y])
```

```python
def  my_rnd_gener(x, p, size = 1000):
    lista = []
    for _ in range(size):
        y = uniform.rvs()
        suma = 0
        for i in range(len(p)):
            if(suma <= y <= suma + p[i]):
                lista.append(x[i])
                suma += p[i]
                break
            suma += p[i]
    return lista

l = my_rnd_gener([0,1,2,3],[0.4,0.2,0.15,0.25])
print(l)
```

```python
from random import randrange
from matplotlib.pyplot import bar, hist, grid, show, legend, xticks
from scipy.stats import binom

x = [-1, 1, 3]
p = [0.35,0.25,0.4]
data = my_rnd_gener(x,p,size = 100000)
bin_edges = [k + 0.5 for k in range(min(x) - 1,max(x) + 1)]
hist(data, bin_edges, density = True, rwidth = 0.9, color = 'blue', edgecolor = 'black', alpha = 0.5, label = 'generator')

bar(x, p, width = 0.85, color = 'red', edgecolor = 'black', alpha = 0.6, label = 'probabilitati')

legend(loc = 'upper left')
grid()
show()
```

```python
from numpy import log

def my_exp_rand(a, size = 1):
    lista = []
    for _ in range(size):
        y = uniform.rvs()
        x = (-1 / a) * log(1-y) 
        lista.append(float(x))
    return lista

l = my_exp_rand(1/5,100)
print(l)
```

```python
from matplotlib.pyplot import plot
from scipy.stats import expon
a = 1/5
data = my_exp_rand(a, size = 100000)
hist(data, bins = 10, density = True, range = (0,30), rwidth = 0.9, color = 'blue', edgecolor = 'black', alpha = 0.5, label = 'generator')

x = range(0,30)
plot(x,expon.pdf(x,loc = 0, scale = 1/a), color = 'red')

xticks(range(0,30,3))
legend(loc = 'upper left')
grid()
show()
```
