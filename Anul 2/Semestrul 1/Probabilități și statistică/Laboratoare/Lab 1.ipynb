```python
from random import sample
from math import factorial
from itertools import permutations, combinations, combinations_with_replacement

# Afișați toate permutările posibile ale cuvântului "word"
list(permutations('word'))
```

```python
# Afișați numărul total al permutărilor cuvântului "word"
len(list(permutations('word')))
```

```python
#Afișați o permutare aleatoare a cuvântului "word"
#sample(list(permutations('word')), 1) - ineficient
sample('word', len('word'))  
```

```python
# Afișați toate aranjamentele posibile ale cuvântului "word"

def aranjamente(word, k):
    return list(permutations(word, k))

aranjamente('word', 2)
```

```python
# Afișați numărul total al aranjamentelor cuvântului "word"

def numar_aranjamente(word, k):
    return len(list(aranjamente(word,k)))
numar_aranjamente('word', 2)
```

```python
#Afișați o combinare aleatoare a cuvântului "word"

def aranjamente_random(word, k):
    return sample(word,k)

aranjamente_random('word',2)
```

```python
# Afișați toate combinarile posibile ale cuvântului "word"

def combinari(word, k):
    return combinations(word, k)
list(combinari('word', 2))
```

```python
# Afișați numărul total al combinărilor cuvântului "word"

def numar_combinari(word, k):
    return len(list(combinations(word, k)))

numar_combinari('word',2)
```

```python
#Afișați o combinare aleatoare a cuvântului "word"

# def combinari_random(word, k):
#     return sample(list(combinations(word, k)),1)

# combinari_random('word',2) - ineficient

def combinari_random(word, k, cate = 2):
    litere = [litera for litera in word]
    rezultat = ''
    lungime = len(word)
    while(cate > 0):
        numereGenerate = list(sample(range(0, lungime), k))
        numereGenerate.sort()
        rezultat += "['"
        rezultat += ''.join([litere[i] for i in numereGenerate])
        cate -= 1
        rezultat += "']"
        rezultat += " "
    return rezultat

combinari_random('word', 2, 5)
    
```
