```python
from scipy.stats import bernoulli

def mers_aleator(n = 100,p = 0.5):
    start = 0
    lista = bernoulli.rvs(p, size=n)
    pozitii = []
    for i in lista:
        start += int(2 * i - 1)
        pozitii.append(start)

    return pozitii

print(mers_aleator())
```

```python
from random import randrange
from matplotlib.pyplot import bar, hist, grid, show, legend, xticks
from scipy.stats import binom

n = 10
p = 0.5
data = [mers_aleator(n, p)[-1] for _ in range(5000)]
bin_edges = [k - 0.5 for k in range(-n, n+2)]
hist(data, bin_edges, density = True, rwidth = 0.9, color = 'green', edgecolor = 'black', alpha = 0.5, label = 'bernoulli')

distribution = dict([2 * k - n, binom.pmf(k, n, p)] for k in range(n + 1))
bar(distribution.keys(), distribution.values(), width = 0.85, color = 'red', edgecolor = 'black', alpha = 0.6, label = 'probabilitati')

legend(loc = 'upper left')
xticks(range(-n, n + 1, 2))
grid()
show()
```
